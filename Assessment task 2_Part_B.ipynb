## Take Appropriate Actions to handle the outliers - Estimate Base Salary

```python
# Define a function to remove outliers based on the IQR method

Q1 = df["Estimate Base Salary"].quantile(0.25)
Q3 = df["Estimate Base Salary"].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
total_outlier_num = ((df["Estimate Base Salary"]<(lower_bound)) | (df["Estimate Base Salary"] > (upper_bound))).sum()

print("Total Number of Outliers in Estimate Base Salary: {}".format(total_outlier_num))
```

```python
def remove_outliers_iqr(column):
    Q1 = column.quantile(0.25)
    Q3 = column.quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return (column >= lower_bound) & (column <= upper_bound)

# Get the boolean mask for the "Estimate Base Salary" column
mask = remove_outliers_iqr(df["Estimate Base Salary"])

# Filter the DataFrame using the mask
df = df[mask]
```

## 6.5.6 Pie Chart 
## What kind of skills required from popular job title Data scientist, Data engineer, Data analyst, Software Engineer

```python
# List of job titles
job_titles = ["Data Scientist", "Data Engineer", "Data Analyst"]

# List of skill columns (assuming they are named as 'python_yn', 'r_yn', etc.)
skill_columns = ['python_yn', 'r_yn', 'sql_yn', 'excel_yn','aws_yn','tableau_yn', 'powerbi_yn']  # Add all skill columns here

for title in job_titles:
    # Filter dataframe for the job title
    filtered_df = df[df["Job Title"] == title]
    
    # Sum the skill columns
    skill_sums = filtered_df[skill_columns].sum().sort_values(ascending=False)
    
    # Plot pie chart
    plt.figure(figsize=(10, 6))
    skill_sums.plot.pie(autopct='%1.1f%%', startangle=140)
    plt.title(f"Skills Required for {title}")
    plt.ylabel('')  # This is to remove the default column name as label
    plt.show()
```

* Data Engineer:
Python & SQL: Large portions for Python (24.3%) and SQL (35.1) are expected as data engineers often write scripts (in Python) and manage large databases (using SQL).
AWS: A significant slice of about 20.3% for AWS suggests that cloud-based data storage and processing services are crucial for data engineering tasks.

* Data Analyst:
Prominent slices for SQL 26.9% and Python 22.7% are expected. Data analysts often use SQL and Python for data retrieval. Tableau & Excel: reasonable portions for these tools indicate that data visualisation and reporting are critical components of a data analyst's role.
